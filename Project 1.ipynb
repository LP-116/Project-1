```python
# Import dependencies 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# CSV file path, converting CSV file into a Data Frame & dropping columns with no values 

csv_file = "Desktop/housing_data.csv"
csv_file_df = pd.read_csv(csv_file)

clean_df = csv_file_df.dropna()
```

```python
# Creating a variables to hold median house prices across all Australian Capital Cities

sydney = clean_df["Median Price of Established House Transfers (Unstratified) ;  Sydney ;"]
melbourne = clean_df["Median Price of Established House Transfers (Unstratified) ;  Melbourne ;"]
brisbane = clean_df["Median Price of Established House Transfers (Unstratified) ;  Brisbane ;"]
perth = clean_df["Median Price of Established House Transfers (Unstratified) ;  Perth ;"]
adelaide = clean_df["Median Price of Established House Transfers (Unstratified) ;  Adelaide ;"]
canberra = clean_df["Median Price of Established House Transfers (Unstratified) ;  Canberra ;"]
hobart = clean_df["Median Price of Established House Transfers (Unstratified) ;  Hobart ;"]
darwin = clean_df["Median Price of Established House Transfers (Unstratified) ;  Darwin ;"]
years = clean_df["Unnamed: 0"]

# Creating a Data Fame

city_data_df = pd.DataFrame({"Sydney": sydney, "Melbourne": melbourne, "Brisbane": brisbane, "Perth": perth, "Adelaide": adelaide,
                         })

# Printing head for sanity check 

city_data_df.head()
```

```python
# Creating line graph plotting median house price over time in Sydney, Melbourne, Brisbane, Perth & Adelaide:

city_data_df.plot(kind="line",figsize=(10,10))
plt.xlabel("Years", fontsize=14)
plt.ylabel("Median House Price ($)", fontsize=14)
plt.title("Major Australian Capital City Median House Price Analysis: 2002 - 2021", fontsize=16)
print("")
```

```python
# Creating variables to hold median house prices across all regional locations in NSW, VIC, QLD, SA & WA

years = years
nsw_regional = clean_df["Median Price of Established House Transfers (Unstratified) ;  Rest of NSW ;"]
vic_regional = clean_df["Median Price of Established House Transfers (Unstratified) ;  Rest of Vic. ;"]
qld_regional = clean_df["Median Price of Established House Transfers (Unstratified) ;  Rest of Qld. ;"]
sa_regional = clean_df["Median Price of Established House Transfers (Unstratified) ;  Rest of SA ;"]
wa_regional = clean_df["Median Price of Established House Transfers (Unstratified) ;  Rest of WA ;"]

# Storing the data in a Data Frame

regional_data_df = pd.DataFrame({"Date": years, "NSW": nsw_regional, "VIC": vic_regional, "QLD": qld_regional, "SA": sa_regional, "WA": wa_regional})

# Creating a line graph plotting the median regional house price over time:

regional_data_df.plot(kind="line",figsize=(10,10))
plt.xlabel("Years", fontsize=14)
plt.ylabel("Median House Price ($)", fontsize=14)
plt.title("NSW, VIC, QLD, SA & WA Regional Median House Price Analysis: 2002 - 2021", fontsize=16)
print("")
```

```python
sydney_median = sydney.median()
melbourne_median = melbourne.median()
brisbane_median = brisbane.median()
perth_median = perth.median()
adelaide_median = adelaide.median()

median_df = pd.DataFrame({sydney_median, melbourne_median, brisbane_median, perth_median, adelaide_median})

median_df["City Name"] = "Melbourne", "Sydney", "Brisbane", "Perth", "Adelaide"

x_axis = median_df["City Name"]
y_axis = median_df[0]

my_colours = ["lightcoral", "gold", "mediumseagreen", "cornflowerblue", "darkorange"]

plt_1 = plt.figure(figsize=(10, 10))
plt.bar(x_axis, y_axis, color=my_colours, align='center') 
plt.title("2002 - 2022: Australian Major Capitals Net Median House Price", fontsize=16)
plt.ylabel("Median House Price (Units = $,000)", fontsize=14)
plt.xlabel("Cities", fontsize=14)
print("")
```
