```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
from scipy import stats
from sklearn import datasets
from matplotlib.ticker import PercentFormatter
import matplotlib.ticker as mtick
```

```python
csv_file_three = "Desktop/growth_rate.csv"
growth_rate_df = pd.read_csv(csv_file_three)
```

```python
x_axis = growth_rate_df['Date']
y_axis_syd = growth_rate_df['Growth Rate - SydneyMedianHouse']
y_axis_melb = growth_rate_df['Growth Rate - MelbMedianHouse']
y_axis_bris = growth_rate_df['Growth Rate - BrisMedianHouse']
y_axis_adelaide = growth_rate_df['Growth Rate - AdelaideMedianHouse']
y_axis_perth = growth_rate_df['Growth Rate - PerthMedianHouse']

plt.figure(figsize=(12,8)) 
plt.scatter(x_axis, y_axis_syd,s=100, label="Sydney")
plt.scatter(x_axis, y_axis_melb,s=100, label="Melbourne")
plt.scatter(x_axis, y_axis_bris,s=100, label="Brisbane")
plt.scatter(x_axis, y_axis_adelaide,s=100, label="Adelaide")
plt.scatter(x_axis, y_axis_perth,s=100, label="Perth")

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis_syd)
regression_values_syd = x_axis * slope + intercept

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis_melb)
regression_values_melb = x_axis * slope + intercept

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis_bris)
regression_values_bris = x_axis * slope + intercept

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis_adelaide)
regression_values_adelaide = x_axis * slope + intercept

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis_perth)
regression_values_perth = x_axis * slope + intercept

regression_value_master = (regression_values_syd + regression_values_melb + regression_values_bris + regression_values_adelaide + regression_values_perth)/5 

plt.xticks(rotation=0, font='Arial', fontsize=13)
plt.yticks(rotation=0, font='Arial', fontsize=13)
plt.title("Capital Cities Median House Price Growth Rate: 2010 - 2020", font='Arial Black', fontsize=18)
plt.xlabel("Year", font='Arial Black', fontsize=16)
plt.ylabel("Growth Rate (%)", font='Arial Black', fontsize=16)
plt.plot(x_axis, regression_value_master,"k")
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.legend(prop={'size': 12})
print("")
```
