### Analysis details:

In this section we looked at 5 different csv files and extracted the total earnings for the years 2017, 2018, 2019 & 2020.
This detail was then merged into a dataset that had the income by state from 2010 to 2016, effectively giving us 10 years worth of data.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Reading in the file required.
vic_salary = "Data files/VIC salary.csv"
vic_salary = pd.read_csv(vic_salary)
```

```python
# Displaying the dataframe.
vic_salary
```

```python
# Setting the list of rows we want to keep in the dataframe.
my_list = [20, 22, 24, 26]

# Filtering the dataframe on the list.
vic_filter = vic_salary[vic_salary.index.isin(my_list)].copy()
vic_filter
```

```python
# Removing any unwanted columns
cols = [1,2,3,4,5,6,7,9]
vic_clean = vic_filter.drop(vic_filter.columns[cols], axis=1)
vic_clean
```

```python
# Renaming the columns.
vic_clean.columns.values[0] = "Period"
vic_clean.columns.values[1] = "Vic salary"
vic_clean
```

```python
# Saving the result to csv file.
vic_clean.to_csv("Data files/vic_clean.csv", encoding="utf-8", index=False, header=True)
```

```python
# Once all states have had a saved csv file we can then merge into 1.
# We start by reading in all the csv files.
vic = pd.read_csv("Data files/vic_clean.csv")
nsw = pd.read_csv("Data files/nsw_clean.csv")
wa = pd.read_csv("Data files/wa_clean.csv")
qld = pd.read_csv("Data files/qld_clean.csv")
sa = pd.read_csv("Data files/sa_clean.csv")
```

```python
# Then we merge 2 together.
combined_salary = pd.merge(vic, nsw, on="Period", how="left")
combined_salary
```

```python
# Then we merge the next one.
combined_salary = pd.merge(combined_salary, wa, on="Period", how="left")
combined_salary 
```

```python
# Then we merge the next one.
combined_salary = pd.merge(combined_salary, qld, on="Period", how="left")
combined_salary 
```

```python
# And then we merge the final one.
combined_salary = pd.merge(combined_salary, sa, on="Period", how="left")
combined_salary 
```

```python
# Here we get rid of the "Nov"
combined_salary["Period"] = combined_salary["Period"].str.replace("Nov-", "")
combined_salary
```

```python
# Lastly, we save the merge dataframe to a csv file so it can be merge with the main dataset.
combined_salary.to_csv("Data files/combined_salary.csv", encoding="utf-8", index=False, header=True)
```
