```python
csv_file = "Desktop/earnings_data.csv"
salary_df = pd.read_csv(csv_file)
salary_df.dropna()
```

```python
syd_wages = salary_df["SydneyAnnualSalary-Median"]
melb_wages = salary_df["MelbourneAnnualSalary-Median"]
bris_wages = salary_df["BrisbaneAnnualSalary-Median"]
adelaide_wages = salary_df["AdelaideAnnualSalary-Median"]
perth_wages = salary_df["PerthAnnualSalary-Median"]
```

```python
x_axis = salary_df["Year"].astype(str)


plt_1 = plt.figure(figsize=(7,5))
plt.bar(x_axis, syd_wages, align='center', label='Sydney', color='dodgerblue')
plt.bar(x_axis, nsw_wages, label='NSW Regional', color='lightskyblue', hatch='/')
plt.title("Sydney v NSW Regional: Median Annual Salary", font='Arial Black', fontsize=18)
plt.xlabel("Year", font='Arial Black', fontsize=16)
plt.ylabel("Median Salary ($)", font='Arial Black', fontsize=16)
plt.legend(prop={'size': 12})
plt.xticks(rotation=0, font='Arial', fontsize=13)
plt.yticks(rotation=0, font='Arial', fontsize=13)
plt.xticks(rotation=0, font='Arial', fontsize=13)
print("")
```

```python
csv_file_two = "Desktop/housing_data.csv"
housing_df = pd.read_csv(csv_file_two)
housing_df.dropna()
```

```python
sydney = housing_df["Median Price of Established House Transfers (Unstratified) ;  Sydney ;"]
melbourne = housing_df["Median Price of Established House Transfers (Unstratified) ;  Melbourne ;"]
brisbane = housing_df["Median Price of Established House Transfers (Unstratified) ;  Brisbane ;"]
perth = housing_df["Median Price of Established House Transfers (Unstratified) ;  Perth ;"]
adelaide = housing_df["Median Price of Established House Transfers (Unstratified) ;  Adelaide ;"]
timeline = housing_df["Date"]

housing_data_df = pd.DataFrame({"Year": timeline, "Sydney": sydney, "Melbourne": melbourne, "Brisbane": brisbane, "Perth": perth, "Adelaide": adelaide,
                         })

housing_data_df.set_index('Year')
```

```python
x_axis = housing_data_df["Year"].astype(str)

plt.figure(figsize=(12,8)) 
plt.plot(x_axis, sydney, color="red", marker='o', label="Sydney") 
plt.plot(x_axis, melbourne, color="seagreen", marker='o',label="Melbourne")
plt.plot(x_axis, brisbane, color="orange", marker='o', label="Brisbane")
plt.plot(x_axis, perth, color="cornflowerblue", marker='o', label="Perth")
plt.plot(x_axis, adelaide, color="purple", marker='o', label="Adelaide")
plt.xticks(rotation=0, font='Arial', fontsize=13)
plt.yticks(rotation=0, font='Arial', fontsize=13)
plt.title("Capital Cities Median House Price: 2010 - 2020", font='Arial Black', fontsize=18)
plt.xlabel("Year", font='Arial Black', fontsize=16)
plt.ylabel("Median House Price ($,000)", font='Arial Black', fontsize=16)
plt.legend(prop={'size': 15})
print("")
```

```python
nsw_regional = housing_df["Median Price of Established House Transfers (Unstratified) ;  Rest of NSW ;"]
vic_regional = housing_df["Median Price of Established House Transfers (Unstratified) ;  Rest of Vic. ;"]
qld_regional = housing_df["Median Price of Established House Transfers (Unstratified) ;  Rest of Qld. ;"]
sa_regional = housing_df["Median Price of Established House Transfers (Unstratified) ;  Rest of SA ;"]
wa_regional = housing_df["Median Price of Established House Transfers (Unstratified) ;  Rest of WA ;"]
```

```python
x_axis = housing_data_df["Year"].astype(str)

plt.figure(figsize=(12,8)) 
plt.plot(x_axis, nsw_regional, color="red", marker='o', label="NSW Regional") 
plt.plot(x_axis, vic_regional, color="seagreen", marker='o',label="VIC Regional")
plt.plot(x_axis, qld_regional, color="orange", marker='o', label="QLD Regional")
plt.plot(x_axis, wa_regional, color="cornflowerblue", marker='o', label="WA Regional")
plt.plot(x_axis, sa_regional, color="purple", marker='o', label="SA Regional")
plt.xticks(rotation=0, font='Arial', fontsize=13)
plt.yticks(rotation=0, font='Arial', fontsize=13)
plt.title("Regional Median House Price: 2010 - 2020", font='Arial Black', fontsize=18)
plt.xlabel("Year", font='Arial Black', fontsize=16)
plt.ylabel("Median House Price ($,000)", font='Arial Black', fontsize=16)
plt.legend(prop={'size': 15})
print("")
```
